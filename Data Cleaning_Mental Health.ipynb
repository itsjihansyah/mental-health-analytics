# Mental Health Dataset Cleaning

## Import Dataset

```python
import pandas as pd
import re
```

```python
df = pd.read_excel(r"D:\ME\Data Analytics Projects\Mental Health\Mental Health Dataset.xlsx")
df
```

## Dropping Duplicates and Unneeded Column

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = "comments")
```

## Replacing and Grouping Categorical Values

```python
df['Gender'].unique()
```

```python
df["Gender"] = df.Gender.apply(lambda s: re.sub(
                                    "((^F)([A-Za-z]+)*)|(^F$)",
                                    "Female",
                                    s.strip().title())
                                )

df["Gender"] = df.Gender.apply(lambda s: re.sub(
                                    "((^M)([A-Za-z]+)*)|(^M$)",
                                    "Male",
                                    s.strip().title())
                                )
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 'Other' if (re.search(r'\bmale\b', x, flags=re.IGNORECASE) and re.search(r'\bfemale\b', x, flags=re.IGNORECASE)) else ('Male' if re.search(r'(\bmale\b)|(\bman\b)|(\bguy\b)', x, flags=re.IGNORECASE) else ('Female' if re.search(r'(\bfemale\b)|(\woman\b)|(\she\b)|(queer\b)', x, flags=re.IGNORECASE) else 'Other')))
```

## Handling Nan Values

```python
df.isna().sum()
```

```python
df[df['state'].isna()==1]
```

```python
df['state'] = df['state'].fillna('Unknown')
```

```python
df['self_employed'] = df['self_employed'].fillna('Unknown')
df['work_interfere'] = df['work_interfere'].fillna('Don\'t know')
```
